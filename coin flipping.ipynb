```python
def coinflip():
    """flips a coin using a random selection of 1 and 2."""
    x = random.randint(1,2)
    if x == 1:
        return "heads"
    else:
        return "tails"
    
def coinflip_n_times(n):
    """flips a coin n times and returns the total number of heads and tails"""
    heads = 0
    tails = 0
    count = 0
    log = 0  # To be used as 2^log
    outcome_dict = {}
    log_list = []
    heads_list = []
    tails_list = []
    ratio_list = []
    difference_list = []
    outcome = [coinflip() for _ in range(n)]  # Flipping coin and storing in a list
    
    for result in outcome:  # Iterating over outcome list and storing the result in heads and tails counters
        count += 1
        if result == "heads":
            heads += 1
        else:
            tails += 1
            
        if count == 2**log:  # storing the count in a dictionary at each point 2^n within n
            #print("the count at {} is heads: {} and tails: {}".format(2**log, heads, tails))\
            log_list.append(2**log)
            heads_list.append(heads)
            tails_list.append(tails)
            difference_list.append(abs(heads - tails))
            
            if tails != 0:
                ratio_list.append(heads/tails)
            else:
                ratio_list.append(tails/heads)
            log += 1 
            
        elif count == n:  # storing the count in a dictionary at point n
            log_list.append(n)
            heads_list.append(heads)
            tails_list.append(tails)
            ratio_list.append(heads/tails)
            difference_list.append(abs(heads - tails))
            
    return log_list, heads_list, tails_list, ratio_list, difference_list
    
```

```python
log_list, heads_list, tails_list, ratio_list, difference_list = coinflip_n_times(100000)
```

```python
def coinflip_n_times_stdev(n):
    """flips a coin n times and returns the total number of heads and tails"""
    heads = 0
    tails = 0
    count = 0
    log = 0  # To be used as 2^log
    log_list = []
    ratio_list = []
    outcome = [coinflip() for _ in range(n)]  # Flipping coin and storing in a list
    
    for result in outcome:  # Iterating over outcome list and storing the result in heads and tails counters
        count += 1
        if result == "heads":
            heads += 1
        else:
            tails += 1
            
        if count == 2**log:  # storing the count at each point 2^n within n
            log_list.append(2**log)
            if tails != 0:  # prevent dividing by zero
                ratio_list.append(heads/tails)
            else:
                ratio_list.append(tails/heads)
            log += 1 
            
        elif count == n:  # storing the count in a dictionary at point n
            log_list.append(n)
            ratio_list.append(heads/tails)
            
    return log_list, ratio_list


def flip_trials(trials, flips):
    """flips a coin x times for y trials and stores the result as a list of lists"""
    n_ratio_list = []
    for i in range(trials):
        log_list, ratio_list = coinflip_n_times_stdev(flips)
        n_ratio_list.append(ratio_list)
    return n_ratio_list
```

```python
n_ratio_list = flip_trials(20, 100000)
```

```python
def coinflip_n_times_heads_ratio(n):
    """flips a coin n times and returns the total number of heads and tails"""
    heads = 0
    tails = 0
    count = 0
    heads_list = []
    outcome = [coinflip() for _ in range(n)]  # Flipping coin and storing in a list
    
    for result in outcome:  # Iterating over outcome list and storing the result in heads and tails counters
        count += 1
        if result == "heads":
            heads += 1
        else:
            tails += 1
            
    return heads/n
    
```

```python
coinflip_n_times_heads_ratio(20)
```

```python
def heads_ratio_list(trials, flips):
    heads_ratio_list = []
    for i in range(trials):
        heads_ratio_list.append(coinflip_n_times_heads_ratio(flips))
    return heads_ratio_list  
```

```python
heads_ratio_list_one_hundred_flips = heads_ratio_list(100000, 100)
```
