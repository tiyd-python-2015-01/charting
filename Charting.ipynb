# Finds the interval values of 2**n

```python
"""Finds the values which will be used as intervals which I plot my heads or tails at."""

nums = []
def find_pow_intervals():
    for n in range(17):
        nums.append(math.pow(2, n))
    #print(nums)
    return nums
print(find_pow_intervals())
```

# Creates a function that represents Heads or Tails as a 1 or 0

```python
"""Creates a function that represents heads or tails as either a 1 or 0"""

def heads_or_tails():
    heads_tails = []
    for n in range(100000):
        flip = random.randint(0, 1)
        heads_tails.append(flip)
        #print(heads_tails)
    return heads_tails
#print(heads_or_tails())
```

# Finds how many heads have been recorded by 2**n

```python
"""Creates a list of numbers, each item in the list represents how many heads there have been by the time it gets to 2**n"""

powers = find_pow_intervals()
h_or_t = heads_or_tails()
headcount = 0
headcount_list = []
powercount = 0
print(len(h_or_t))

for n in range(100000):
    if h_or_t[n] == 1:
        headcount += 1
    if n == 2**(powercount):
        headcount_list.append(headcount)
        powercount += 1
print(headcount_list)
```

```python
def coin_flip(num):
    heads = 0
    tails = 0
    for i in range(0, num):
        flip = random.randint(0, 1)
        if (flip == 0):
            heads += 1
        else:
            tails += 1
    return [heads, tails]
coin_flip(10000)
flip_results = coin_flip(100)

flip_ratio = flip_results[0]/flip_results[1]
print(flip_ratio)
```
