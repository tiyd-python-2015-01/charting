```python
import random
import statistics as st
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def flip_coin():
    """If > .5 Heads, else Tails"""
    return random.random() > .5

def flip_coin_n_times(n=100000):
    """Flips coin n-times and returns list of counts at intervals of 2^n flips"""
    heads = 1
    tails = 1
    record_count = 0
    data = []
    
    for flip_count in range(n):
        if flip_coin():
            heads += 1
        else:
            tails += 1
        if flip_count == 2**record_count:
            data.append((heads, tails))
            record_count += 1
    data.append((heads,tails))
    return data
```

```python
"""Flip coins 100000 times and find the difference"""
data = flip_coin_n_times(100000)
difference = [abs(flip[0] - flip[1]) for flip in data]
```

```python
"""Set up and show the plot of difference"""
x1 = ["2^{}".format(n) for n in range(len(data))]
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.plot(difference)
plt.xticks(range(len(x1)), x1)
plt.xlabel("Number of Flips")
plt.ylabel("Heads - Tails")
plt.title("Difference Between Heads and Tails")
plt.show()
```

```python
"""Calculate the ratio of heads to tails from data"""
ratio = [flip[0]/flip[1] for flip in data]
```

```python
"""Set up and show the plot for ratio"""
x1 = ["2^{}".format(n) for n in range(len(data))]
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.plot(ratio)
plt.xticks(range(len(x1)), x1)
plt.xlabel("Number of Flips")
plt.ylabel("Heads / Tails")
plt.title("Ratio Between Heads to Tails")
plt.show()
```

From the above graphs we can see that as the sample size increases the size of the random swings may increase and heads or tails may occur hundreds more time, but in relation to the total number of iterations these swings are small.  As the sample size increases the ratio of heads to tails approaches 1.

```python
"""Set up and show scatter plot of difference using a log scale for the x-axis"""
x2 = [2**n for n in range(len(data))]
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.xscale("log")
plt.xlabel("Number of Flips")
plt.ylabel("Heads - Tails")
plt.title("Difference Between Heads to Tails")
plt.scatter(x2, difference)
plt.show()
```

```python
"""Set up and show a scatter plot with log scale x-axis for ratio"""
x2 = [2**n for n in range(len(data))]
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.xscale("log")
plt.xlabel("Number of Flips")
plt.ylabel("Heads / Tails")
plt.title("Ratio of Heads to Tails")
plt.scatter(x2, ratio)
plt.show()
```

```python
"""Run 20 simulations of 100000 iterations, rotate the 2d-array and find the mean of each row.  Store data in a list
of means"""

data_list = []
ratio_list = []
mean_list = []

for count in range(20):
    data = flip_coin_n_times(100000)
    data_list.append(data)
    
data_list = list(zip(*data_list[::-1]))

for row in data_list:
    ratio_list.append([item[0]/item[1] for item in row])

for row in ratio_list:
    mean_list.append(st.mean(row))
```

```python
"""Set up and show the scatter plot for the mean data using a log scale for the x-axis"""
x2 = [2**n for n in range(len(data))]
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.xscale("log")
plt.xlabel("Number of Flips")
plt.ylabel("Mean of Ratios of Heads / Tails")
plt.title("Means of Ratios of Heads to Tails")
plt.scatter(x2, mean_list)
plt.show()
```

```python
"""Calculate the standard deviation of the ratio of heads to tails for each of the 20 simulations run in the previous section"""
std_dev_list = []
for row in ratio_list:
    std_dev_list.append(st.stdev(row))
```

```python
"""Set up and display the scatter plot for the standard deviation data"""
x2 = [2**n for n in range(len(data))]
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Number of Flips")
plt.ylabel("Standard Deviation of Ratio of Heads / Tails")
plt.title("Standard Deviations of Ratios of Heads to Tails")
plt.scatter(x2, std_dev_list)
plt.show()
```

The scatter plots show that at low sample size the mean of the data may seem to swing wildly, but at higher sample sizes the data equalizes an the mean settles almost exactly around one.  The standard deviation graph shows that as sample size increases the standard deviation decreases linearly.

```python
"""Run 100000 simulations of 100 coin flips.  Calculate the ratio of heads to total flips"""
data_list = []
head_ratio = []

for count in range(100000):
    data = flip_coin_n_times(100)
    data_list.append(data)

for trial in data_list:
    head_ratio.append(trial[len(trial)-1][0]/100)
```

```python
"""Calculate the mean and standard deviation of the previous simulations"""
mean = st.mean(head_ratio)
std_dev = st.stdev(head_ratio)
```

```python
"""Set up and display the histogram of the ratio of Heads results to total flips for the previous simulation."""
plt.hist(head_ratio, bins=40)
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.yscale("log")
plt.xlabel("Trial")
plt.ylabel("Head/Total Ratio")
plt.title("Head/Total Ratio Histogram\n"
           "Mean = {}, Sandard Deviation = {}".format(mean, std_dev))

ymin, ymax = plt.ylim()
plt.show()
```

```python
"""Run 100000 simulations of 1000 coin flips.  Calculate the ratio of heads to total flips for each simulation."""
data_list = []
head_ratio2 = []

for count in range(100000):
    data = flip_coin_n_times(1000)
    data_list.append(data)

for trial in data_list:
    head_ratio2.append(trial[len(trial)-1][0]/1000)
```

```python
"""Calculate the mean and standard deviation of the previous simulations."""
mean = st.mean(head_ratio2)
std_dev = st.stdev(head_ratio2)
```

```python
"""Set up and display the histgram for the previous simulations."""
plt.hist(head_ratio2, bins=40)
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.yscale("log")
plt.xlabel("Trial")
plt.ylabel("Head/Total Ratio")
plt.title("Head/Total Ratio Histogram\n"
           "Mean = {}, Sandard Deviation = {}".format(mean, std_dev))

ymin, ymax = plt.ylim()
plt.show()
```

```python
"""Set up and display the box plot comparing the last two simulations."""
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)
plt.boxplot([head_ratio, head_ratio2])
ymin, ymax = plt.ylim()
plt.ylim(ymin - .1, ymax + .1)
plt.title("Proportion of Heads in Simulations")
plt.ylabel("Proportion of Heads")
plt.xticks(range(1,3), ["100/100000", "1000/100000"])
plt.show()
```

The histograms show that the results of coin flips have a normal distribution.  Most of the runs of coin flips fall very close to a 50% ratio of heads.  Again, a tenfold increase in sample size shows a dramatic decrease in the variability of the simulation.
