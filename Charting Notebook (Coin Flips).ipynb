```python
import random
import math
import statistics as st
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
def flip_coin():
    heads_or_tails = ['H', 'T']
    return random.choice(heads_or_tails)
```

```python
heads = []
tails = []
interval_difference_counts = []
heads_count = []
tails_count = []

def trial_flip_n_times(n=100000):
    count = 0
    i = 0
    results = [flip_coin() for _ in range(n)]
    for result in results:
        count += 1
        if result == 'H':
            heads.append(result)
        else:
            tails.append(result)
        if count == 2 ** (i):
            heads_count.append(len(heads))
            tails_count.append(len(tails))
            interval_difference_counts.append(abs(len(heads) - len(tails)))
            i += 1
            
            

trial_flip_n_times()
print("Heads: {}".format(len(heads)))
print("Tails: {}".format(len(tails)))
print("Heads at 2**n: {}".format(heads_count))
print("Tails at 2**n: {}".format(tails_count))
print("Difference at 2 ** n: {}".format(interval_difference_counts))

        
    
```

```python
heads = []
tails = []
heads_count = []
tails_count = []
means = []
stdevs = []

def ratio_of_lists(list1, list2):
    ratios = [round(x/y, 2) for x, y in zip(list1,list2)]
    return ratios

for _ in range(20):
    trial_flip_n_times()
    means.append(st.mean(ratio_of_lists(heads_count, tails_count)))
    stdevs.append(st.pstdev(ratio_of_lists(heads_count, tails_count)))
    heads_count = []
    tails_count = []
    
print(means)
print(stdevs)    
   
    
    
```
