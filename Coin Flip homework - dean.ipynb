```python
import random
import statistics as st
import matplotlib.pyplot as plt
import math
import seaborn as sb
from operator import sub
import numpy as np
```

```python
%matplotlib inline
```

```python
sides = [0, 1]

def coin_flip():
    return random.choice(sides)

coin_flip()
```

```python
"""Generate list of total number of head results at x intervals."""

def flip_counts(n):
    tails = []
    bigger_list = []
    xs = [pow(2, exp) for exp in range(n) if pow(2, exp) < n]
    xs.append(n)
    counter = 0
    while counter <= n:
        coin_flip()
        if coin_flip() == 1:  
            tails.append(1)
        if counter in xs:
            temp_list = tails[:]
            bigger_list.append(temp_list)
        counter += 1

    ys = [len(i) for i in bigger_list]
    
    return xs, ys



flip_counts(100000)
```

```python

x1 = np.array([1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 100000])
ys = np.array([1, 1, 2, 5, 7, 15, 29, 63, 128, 241, 500, 1016, 2057, 4086, 8259, 16414, 32728, 49956])
y1 = x1 - ys


plt.plot(y1)
plt.xticks(range(19), x1)
plt.xlabel("Powers of 2")
plt.ylabel("Difference")
plt.title("Heads vs Tails")
plt.show()
```

```python
"""20 trials of 100000 flips"""
sides = [0, 1]

def coin_flip():
    return random.choice(sides)

coin_flip()

def flip_counts(n):
    tails = []
    bigger_list = []
    xs = [pow(2, exp) for exp in range(n) if pow(2, exp) < n]
    xs.append(n)
    counter = 0
    while counter <= n:
        coin_flip()
        if coin_flip() == 1:  
            tails.append(1)
        if counter in xs:
            temp_list = tails[:]
            bigger_list.append(temp_list)
        counter += 1

    ys = [len(i) for i in bigger_list]
    
    return xs, ys

for i in range(20):
    flip_counts(100000)
```

```python
xs, ys = flip_counts(10000)
```

```python
ys = np.array(ys)
x1 = np.array(xs)
y1 = x1 - ys

plt.scatter(x1, y1)
plt.xticks(range(19), x1)
plt.xscale('log')
plt.xlabel("Powers of 2")
plt.ylabel("Difference")
plt.title("Heads vs Tails")
plt.show()
```
