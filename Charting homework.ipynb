```python
def flip_coin():
    if random.random() > 0.5:
        return "Head"
    else:
        return "Tail"
```

```python
def flip_coins(number_of_coins):
    counter = 1
    interval_counter = 0
    flips_record = {"Head" : 0, "Tail": 0}
    interval_record = []
    while number_of_coins > counter:
        coin = flip_coin()      
        if coin in flips_record:
            flips_record[coin] += 1
        else:
            flips_record[coin] = 1
        if ((2 ** interval_counter) == counter):
            flips_record["difference"] = abs(flips_record['Head'] - flips_record['Tail'])  
            if flips_record["Head"] == 0 or flips_record["Tail"] == 0:
                flips_record['ratio'] = 0
            else:
                flips_record['ratio'] = flips_record['Head'] / abs(flips_record['Tail'])
            flips_record['ratio'] = round(flips_record['ratio'], 4)
            interval_record.append(flips_record.copy())
            interval_counter += 1
        counter += 1
    interval_record.append(flips_record.copy())
    return interval_record
```

```python
flips_interval_list = flip_coins(100000)
flips_interval_list_two = flip_coins(100000)
flips_interval_list_three = flip_coins(100000)
```

# 20 Trials with 100,000 coin flips each.

```python
def flipping_coin_trials(number_of_trials, number_of_flips):
    list_of_trials = []
    counter = 0
    while number_of_trials > counter:
        list_of_trials.append(flip_coins(number_of_flips))
        counter += 1
    return list_of_trials
```

```python
flips_trial_list = flipping_coin_trials(20, 100000)
flips_trial_list_second = flipping_coin_trials(20, 100000)
flips_trial_list_third = flipping_coin_trials(20, 100000)
```

```python

def make_list_of_ratio_lists(list_of_lists):
    list_ratio_trials = []
    for list_dicts in flips_trial_list:
       list_ratio_trials.append([dicts["ratio"] for dicts in list_dicts])
    each_ratio_list = list(zip(*list_ratio_trials[::-1]))
    return each_ratio_list

def mean_of_ratio_lists(each_ratio_list):
    ratios_means_list= []
    for ratios_lists in each_ratio_list:
        ratios_means_list.append(round(st.mean(ratios_lists), 4))
    return ratios_means_list

def stdev_of_ratio_lists(each_ratio_list):
    ratios_means_list= []
    for ratios_lists in each_ratio_list:
        ratios_means_list.append(round(st.pstdev(ratios_lists), 4))
    return ratios_means_list

ratios_lists = make_list_of_ratio_lists(flips_trial_list)
mean_list = mean_of_ratio_lists(ratios_lists)
stdev_list = stdev_of_ratio_lists(ratios_lists)
```

# 100,000 Trials with 100 flips each

```python
def heads_flip_ratio(number_of_trials, number_of_coins):
    coin_counter = 0
    trial_counter = 0
    ratio_list = []
    while number_of_trials > trial_counter:
        coin_counter = 0
        heads_counter = 0
        while number_of_coins > coin_counter:
            flipped_coin = flip_coin()
            if flipped_coin == "Head":
                heads_counter += 1
            coin_counter += 1
        ratio = heads_counter/number_of_coins
        ratio_list.append(ratio)
        trial_counter += 1
    return ratio_list
```

```python
big_list_of_ratios = heads_flip_ratio(100000, 100)
big_list_of_ratios_two = heads_flip_ratio(100000, 100)
big_list_of_ratios_three = heads_flip_ratio(100000, 100)
```
