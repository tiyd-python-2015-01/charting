###### My simple coin flip function:

```python
def coin_flip():
    x = random.randint(0,1)
    if x == 0:
        return 'heads'
    else:
        return 'tails'
```

###### My multiple flips function:

```python
def flipper(n=100000):
    """flips a coin 100,000 times.  Returns a tuple of the 2**n markers, the breakdown of head and tails,
    a list of the differences and a list of the ratios of x to y"""
    flips = []
    counts = []
    fliplist = []
    ratios = []
    heads = [x for x in flips if x=='heads']
    tails = [x for x in flips if x=='tails']
    for x in range(1,n):
        flips.append(coin_flip())
    #print(flips)
    for y in range (0,n):
        if 2**y < n:
            counts.append(2**y)
            fliplist.append(flips[:2**y])
    counts.append(n)
    fliplist.append(flips)
    headtails = [(x.count('heads'), x.count('tails')) for x in fliplist]
    difflist = [x[0]-x[1] for x in headtails]
    def ratio(x, y):
        if y > 0:
            return x/y
        else:
            return 0
    ratiolist =[ratio(x[0], x[1]) for x in headtails]
     
    return counts, headtails, difflist, ratiolist
```

```python
counts, headtails, difflist, ratiolist = flipper()
```

```python
print(counts)
print(headtails)
print(difflist)
print(ratiolist)
```

###### You can see in the early stages that after 8, 16, even 32 flips, it had been really imbalanced towards heads.  More than you'd expect even in such a small sample size.  So it's fun to see especially in the ratio graph of this one, how much of a spike that creates
