```python
import random
import statistics
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def coin_flip():
    """ Heads == True, Tails == False"""
    return random.randint(0, 1) == 0


def flip_a_lot():
    """ Returns a tuple. [0] is the entire list of coin tosses.
    [1] is the recorded stats at each interval"""
    flips = []
    intervals = []
    interval = 0
    for n in range(100000):
        flips.append(coin_flip())
        if n == 2**interval:
            intervals.append((flips.count(True), flips.count(False)))
            interval += 1
    return flips, intervals
```

```python
entire_list, interval_stats = flip_a_lot()
```

```python
"""Takes a list of trials and returns a list where each
list is an interval from a trial.
[ [1st, 2nd, 3rd], [1st, 2nd, 3rd] ] = [[ 1st, 1st], [2nd, 2nd], [3rd, 3rd]]"""


def create_list_of_each_interval(list_of_trials):
    list_of_intervals = []
    for trial_number in range(len(list_of_trials[0])):
        list_from_trial_of_an_interval = []
        for trial in list_of_trials:
            list_from_trial_of_an_interval.append(trial[trial_number])
        list_of_intervals.append(list_from_trial_of_an_interval)
    return list_of_intervals

```

```python
"""Takes a list of each interval set from the 20 trials 
and takes a difference average.  Returns a list with each 
average corresponding to the interval. """


def create_list_of_averages(list_of_intervals):
    list_of_averages = []
    for interval in list_of_intervals:
        list_of_ratios = []
        for trial in interval:
            if trial[1] == 0:
                list_of_ratios.append(abs(trial[0] / 1))
            else:
                list_of_ratios.append(abs(trial[0] / trial[1]))
        list_of_averages.append(statistics.mean(list_of_ratios))
    return list_of_averages

def create_list_of_stdev(list_of_intervals):
    list_of_stdev = []
    for interval in list_of_intervals:
        list_of_ratios = []
        for trial in interval:
            if trial[1] == 0:
                list_of_ratios.append(abs(trial[0] / 1))
            else:
                list_of_ratios.append(abs(trial[0] / trial[1]))
        list_of_stdev.append(statistics.stdev(list_of_ratios))
    return list_of_stdev
```

```python
intervals = create_list_of_each_interval(list_of_trials)
averages = create_list_of_averages(intervals)
deviations = create_list_of_stdev(intervals)
```

```python
def flips(num_of_flips):
    """Returns a tuple. [0] is the entire list of coin tosses. 
    [1] is the recorded stats at each interval"""
    flips = []
    for n in range(num_of_flips):
        flips.append(coin_flip())
    return flips
```

```python
def trials(num_of_trials, num_of_flips):
    trials = []
    for _ in range(num_of_trials):
        trials.append(flips(num_of_flips))
    return trials
```

```python
ratios_100 = [trial.count(True)/100 
              for trial in trials(100000, 100)]
```

```python
plt.hist(ratios_100, 20)
plt.show()
```

```python
ratios_1000 = [trial.count(True)/1000 
               for trial in trials(10000, 1000)]
```

```python
plt.hist(ratios_1000, 40)
plt.show()
```
